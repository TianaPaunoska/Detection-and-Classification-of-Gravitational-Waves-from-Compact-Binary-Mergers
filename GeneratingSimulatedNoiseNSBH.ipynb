```python
import matplotlib.pyplot as pp
import pycbc.noise
import pycbc.psd
import numpy as np
import os
import zipfile
from google.colab import files
import pandas as pd
import random
```

```python
snr1 = [4,5,6,7]
snr2 = [8,9,10,11]
snr3 = [12,13,14,15]
snr4 = [16,17,18,19]
snr5 = [20,21,22,23]
snr6 = [24,25,26,27]
snr7 = [28,29,30,4]
snr_values = np.arange(4, 31)
flow_array = [10,15,20,25,30]
```

```python
output_dir = 'noise_nsbh_samples'
os.makedirs(output_dir, exist_ok=True)
```

```python
group1 = np.arange(1, 1666)
group2 = np.arange(1666, 3326)
group3 = np.arange(3326, 4986)
group4 = np.arange(4986, 6646)
group5 = np.arange(6646, 8306)
group6 = np.arange(8306, 9966)
group7 = np.arange(9966, 11626)
files_total = np.arange(1, 11626)
group8 = np.arange(1, 3501)
```

```python
noise_nsbh_samples = []
```

```python
for i in group1:
  nsbh_waveform_file = f'/content/NSBH/NSBH/strain{i}.npy'
  nsbh_waveform = np.load(nsbh_waveform_file)
  for snr in snr1:
    flow = random.choice(flow_array)
    delta_f = 1.0 / 16
    flen = int(2048 / delta_f) + 1
    psd = pycbc.psd.aLIGOZeroDetHighPower(flen, delta_f, flow)

    desired_std = np.max(np.abs(nsbh_waveform)) / snr

    delta_t = 1.0 / 2048
    tsamples = len(nsbh_waveform)
    noise = pycbc.noise.noise_from_psd(tsamples, delta_t, psd, seed=np.random.randint(0, 2**32))
    noise *= desired_std / np.std(noise)

    mixed_signal = nsbh_waveform + noise

    noise_nsbh_samples.append(mixed_signal)
```

```python
output_dir = 'noise_nsbh_samples'
os.makedirs(output_dir, exist_ok=True)

file_paths = []
labels = []

for i, noise_nsbh_sample in enumerate(noise_nsbh_samples):

    folder_name = f'Noise_NSBH{i + 1}'
    folder_path = os.path.join(output_dir, folder_name)
    os.makedirs(folder_path, exist_ok=True)

    # zachuvuvanje na sekoj primerok kako npy file
    file_path = os.path.join(folder_path, f'noise_nsbh{i + 1}.npy')
    np.save(file_path, noise_nsbh_sample)

    file_paths.append(file_path)
    labels.append('Noise+NSBH')

label_array = np.column_stack((file_paths, labels))

# zachuvuvanje na labelite za sekoj primerok kako npy file
label_file_path = os.path.join(output_dir, 'labels_noise_nsbh.npy')
np.save(label_file_path, label_array)
```

```python
zip_file_path = 'noise_nsbh_samples.zip'
with zipfile.ZipFile(zip_file_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for file_path in file_paths:
        zipf.write(file_path)


files.download(zip_file_path)
```
