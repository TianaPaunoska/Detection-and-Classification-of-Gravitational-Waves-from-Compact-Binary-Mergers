Генерирање на labelsnoisebns.npy во кој ќе се чуваат патеките на секој noise+bns sample npy file придружени со колона за лабелата 'Noise+BNS'

```python
import numpy as np

num_samples = 50000

file_paths = [f"/kaggle/input/noisebnssamples/Noise+BNS/noise_bns{i}.npy" for i in range(1, num_samples + 1)]

labels = np.array(['Noise+BNS'] * num_samples)

data = np.array(list(zip(file_paths, labels)), dtype=[('Path', 'U100'), ('Label', 'U10')])

np.save('labelsnoisebns.npy', data)
```

```python
import pandas as pd

data = np.load('labelsnoisebns.npy')

df = pd.DataFrame(data)

print(df.head())
```
