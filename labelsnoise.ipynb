Генерирање на labelsnoise.npy во кој ќе се чуваат патеките на секој noise sample npy file придружени со колона за лабелата 'Noise'

```python
import numpy as np

num_samples = 50000

file_paths = [f"/kaggle/input/noisesamples/Noise/noise{i}.npy" for i in range(1, num_samples + 1)]

labels = np.array(['Noise'] * num_samples)

data = np.array(list(zip(file_paths, labels)), dtype=[('Path', 'U100'), ('Label', 'U10')])

np.save('labelsnoise.npy', data)
```

```python
import pandas as pd

data = np.load('labelsnoise.npy')

df = pd.DataFrame(data)

print(df.head())
```
