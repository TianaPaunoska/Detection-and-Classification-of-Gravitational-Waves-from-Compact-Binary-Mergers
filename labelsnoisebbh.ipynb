Генерирање на labelsnoisebbh.npy во кој ќе се чуваат патеките на секој noise+bbh sample npy file придружени со колона за лабелата 'Noise+BBH'

```python
import numpy as np

num_samples = 50000

file_paths = [f"/kaggle/input/noisebbhsamples/Noise+BBH/noise_bbh{i}.npy" for i in range(1, num_samples + 1)]

labels = np.array(['Noise+BBH'] * num_samples)

data = np.array(list(zip(file_paths, labels)), dtype=[('Path', 'U100'), ('Label', 'U10')])

np.save('labelsnoisebbh.npy', data)
```

```python
import pandas as pd

data = np.load('labelsnoisebbh.npy')

df = pd.DataFrame(data)

print(df.head())
```
