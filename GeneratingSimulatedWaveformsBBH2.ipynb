```python
import pylab
from pycbc.catalog import Merger
from pycbc.filter import resample_to_delta_t, highpass
import pycbc.catalog
from pycbc.types import TimeSeries
import numpy as np
import pandas as pd
import os
from pycbc.waveform import get_td_waveform
from pycbc.psd import interpolate, inverse_spectrum_truncation
from pycbc.filter import matched_filter
from pycbc.filter import sigma
```

```python
spins = []

start_spin = -0.8
end_spin = 0.8

step = 0.4

current_spin = start_spin
while current_spin <= end_spin:
    spins.append(current_spin)
    current_spin += step
```

```python
templates_BBH_1 = []
```

```python
m1 = 6
for m2 in range(m1, 101, 5):
  if m1 >= m2:
    mass_ratio = m1/m2
  else:
    mass_ratio = m2/m1
  if mass_ratio <= 10:
    for spin1 in spins:
      for spin2 in spins:
        hp, hc = get_td_waveform(
            approximant="TaylorF2",
            mass1=m1,
            mass2=m2,
            spin1x=0,
            spin1y=0,
            spin1z=spin1,
            spin2x=0,
            spin2y=0,
            spin2z=spin2,
            delta_t=1.0/2048,
            f_lower=10.0
            )
        waveform = hp.numpy()
        templates_BBH_1.append(waveform)
```

```python
print(len(templates_BBH_1))
```

```python
desired_length = 2048

modified_waveforms = []

for waveform in templates_BBH_1:
    if len(waveform) > desired_length:
        modified_waveform = waveform[len(waveform) - desired_length:]
        modified_waveforms.append(modified_waveform)
    else:
        modified_waveforms.append(waveform)

templates_BBH_1 = modified_waveforms
```

```python
print(len(templates_BBH_1[0]))
```

```python
output_dir = '/kaggle/working/BBH_events'
os.makedirs(output_dir, exist_ok=True)

event_paths_1 = []
event_targets_1 = []

i = 5000

for waveform in templates_BBH_1:
  event_name = f'BBH{i + 1}'

  event_dir = os.path.join(output_dir, event_name)
  os.makedirs(event_dir, exist_ok=True)

  np.save(os.path.join(event_dir, f'strain{i + 1}.npy'), waveform)

  event_paths_1.append(os.path.join(event_dir, f'strain{i + 1}.npy'))
  event_targets_1.append('BBH')

  i = i+1

labels = np.array([event_paths_1, event_targets_1]).T
np.save(os.path.join(output_dir, 'labelsnew1.npy'), labels)
```

```python
import shutil

shutil.make_archive("/kaggle/working/BNS_events", 'zip', output_dir)
```
