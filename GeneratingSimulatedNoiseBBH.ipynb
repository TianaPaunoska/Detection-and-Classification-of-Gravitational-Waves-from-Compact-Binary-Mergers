```python
import matplotlib.pyplot as pp
import pycbc.noise
import pycbc.psd
import numpy as np
import os
import zipfile
from google.colab import files
import pandas as pd
import random
```

```python
noise_bbh_samples = []
```

```python
for i in group1:
  bbh_waveform_file = f'/content/BBH/BBH/strain{i}.npy'
  bbh_waveform = np.load(bbh_waveform_file)
  for snr in snr1:
    flow = random.choice(flow_array)
    delta_f = 1.0 / 16
    flen = int(2048 / delta_f) + 1
    psd = pycbc.psd.aLIGOZeroDetHighPower(flen, delta_f, flow)

    desired_std = np.max(np.abs(bbh_waveform)) / snr

    delta_t = 1.0 / 2048
    tsamples = len(bbh_waveform)
    noise = pycbc.noise.noise_from_psd(tsamples, delta_t, psd, seed=np.random.randint(0, 2**32))
    noise *= desired_std / np.std(noise)

    mixed_signal = bbh_waveform + noise

    noise_bbh_samples.append(mixed_signal)

pp.plot(noise_bbh_samples[200].sample_times, noise_bbh_samples[200])
pp.ylabel('Strain')
pp.xlabel('Time (s)')
pp.show()
```

```python
output_dir = 'noise_bbh_samples'
os.makedirs(output_dir, exist_ok=True)

file_paths = []
labels = []

for i, noise_bbh_sample in enumerate(noise_bbh_samples):

    folder_name = f'Noise_BBH{i + 1}'
    folder_path = os.path.join(output_dir, folder_name)
    os.makedirs(folder_path, exist_ok=True)

    # zachuvuvanje na sekoj primerok kako npy file
    file_path = os.path.join(folder_path, f'noise_bbh{i + 1}.npy')
    np.save(file_path, noise_bbh_sample)

    file_paths.append(file_path)
    labels.append('Noise+BBH')

label_array = np.column_stack((file_paths, labels))

# zachuvuvanje na labelite za sekoj primerok kako npy file
label_file_path = os.path.join(output_dir, 'labels_noise_bbh.npy')
np.save(label_file_path, label_array)
```

```python
zip_file_path = 'noise_bbh_samples.zip'
with zipfile.ZipFile(zip_file_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for file_path in file_paths:
        zipf.write(file_path)


files.download(zip_file_path)
```
