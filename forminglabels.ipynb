Спојување на одделните npy фајлови од класите со лабели во еден фајл labels.npy

```python
import numpy as np

data_noise = np.load('/kaggle/input/labels/labelsnoise.npy')
data_bbh = np.load('/kaggle/input/labels/labelsnoisebbh.npy')
data_bns = np.load('/kaggle/input/labels/labelsnoisebns.npy')
data_nsbh = np.load('/kaggle/input/labels/labelsnoisensbh.npy')

merged_data = np.concatenate((data_noise, data_bbh, data_bns, data_nsbh))

np.save('labels.npy', merged_data)
```

```python
import pandas as pd

data = np.load('labels.npy')

df = pd.DataFrame(data)

print(df.head())
```
