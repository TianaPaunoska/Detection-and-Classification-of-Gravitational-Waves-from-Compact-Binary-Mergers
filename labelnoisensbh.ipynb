Генерирање на labelsnoisensbh.npy во кој ќе се чуваат патеките на секој noise+nsbh sample npy file придружени со колона за лабелата 'Noise+NSBH'

```python
import numpy as np

num_samples = 50000

file_paths = [f"/kaggle/input/noisensbhsamples/Noise+NSBH/noise_nsbh{i}.npy" for i in range(1, num_samples + 1)]

labels = np.array(['Noise+NSBH'] * num_samples)

data = np.array(list(zip(file_paths, labels)), dtype=[('Path', 'U100'), ('Label', 'U10')])

np.save('labelsnoisensbh.npy', data)
```

```python
import pandas as pd

data = np.load('labelsnoisensbh.npy')

df = pd.DataFrame(data)

print(df.head())
```
